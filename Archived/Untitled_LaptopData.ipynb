```python
df_by_Payment_Mode=df.groupby('Payment_Mode')['Customer_ID'].count()
print(df_by_Payment_Mode)
df_by_Payment_Mode.plot(kind='bar', title='Payment_Mode wise Count')
```

```python
df_by_Brand_Purchased=df.groupby('Brand_Purchased')['Customer_ID'].count()
print(df_by_Brand_Purchased)
df_by_Brand_Purchased.plot(kind='bar', title='Brand_Purchased wise Count')
```

```python
df_by_Purchase_Mode=df.groupby('Purchase_Mode')['Customer_ID'].count()
print(df_by_Purchase_Mode)
df_by_Purchase_Mode.plot(kind='bar', title='Purchase_Mode wise Count')
```

```python
df_by_city=df.groupby('City')['Customer_ID'].count()
print(df_by_city)
df_by_city.plot(kind='bar', title='City wise Count')
```
