**Reasoning**:
Identify potential outliers in numerical columns using the IQR method and visualize distributions and relationships using plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

def identify_outliers_iqr(df, column):
    """Identifies outliers using the IQR method."""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

def analyze_dataframe(df, df_name):
    """Performs outlier detection and visualization for a DataFrame."""
    print(f"--- Analyzing Outliers and Visualizing Distributions for {df_name} ---")
    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

    outliers_dict = {}
    for col in numerical_cols:
        outliers = identify_outliers_iqr(df, col)
        if not outliers.empty:
            outliers_dict[col] = outliers
            print(f"Potential outliers found in '{col}': {len(outliers)}")

    if outliers_dict:
        print(f"Storing outliers for {df_name} in 'outliers_{df_name}' variable.")
        globals()[f'outliers_{df_name}'] = outliers_dict
    else:
        print(f"No significant outliers found in {df_name} using IQR method.")

    # Visualize distributions
    print(f"\n--- Visualizing Distributions for {df_name} ---")
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(numerical_cols):
        plt.subplot(int(len(numerical_cols)/4) + 1, 4, i + 1)
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(col)
    plt.tight_layout()
    plt.show()

    # Visualize relationships (example: with 'price' for train/test, 'loan_status' for credit_risk)
    print(f"\n--- Visualizing Relationships for {df_name} ---")
    if df_name in ['train_df', 'test_df']:
        if 'price' in numerical_cols:
            plt.figure(figsize=(15, 10))
            for i, col in enumerate([c for c in numerical_cols if c != 'price']):
                 plt.subplot(int(len(numerical_cols)/4) + 1, 4, i + 1)
                 sns.scatterplot(data=df, x=col, y='price')
                 plt.title(f'{col} vs Price')
            plt.tight_layout()
            plt.show()
    elif df_name == 'credit_risk_df':
         if 'loan_status' in df.columns:
            numerical_cols_credit = df.select_dtypes(include=['float64', 'int64']).columns.drop('loan_status', errors='ignore')
            plt.figure(figsize=(15, 10))
            for i, col in enumerate(numerical_cols_credit):
                 plt.subplot(int(len(numerical_cols_credit)/4) + 1, 4, i + 1)
                 sns.boxplot(data=df, x='loan_status', y=col)
                 plt.title(f'{col} vs Loan Status')
            plt.tight_layout()
            plt.show()


analyze_dataframe(train_df, 'train_df')
analyze_dataframe(test_df, 'test_df')
analyze_dataframe(credit_risk_df, 'credit_risk_df')
```
